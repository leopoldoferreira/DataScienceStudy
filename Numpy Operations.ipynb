# Numpy Operations 
- Aritmetics operations between arrays
- Operations between scalar arrays
- Some pre programmed functions of numpy

```python
import numpy as np
```

```python
arr = np.arange(0,16)
```

```python
arr
```

```python
arr + arr # add arrays *should be the same size
arr - arr
arr * arr
```

```python
arr / arr #if your divide by 0 the code stops and output only an error, numpy output only an warning and display "nan"
```

```python
1 / arr # return inf
```

```python
arr + 100 # add the number on all elements
arr - 100
arr ** 10
arr * 100
arr / 100 
```

```python
np.sqrt(arr) # return square root of each elements
np.exp(arr) # return exponencial of each elements
np.mean(arr) # return the mean of all elements
np.std(arr) # return the standard deviation of all elements
np.sin(arr) # return the sine of each element
```
