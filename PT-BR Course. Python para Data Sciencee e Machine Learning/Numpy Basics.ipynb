```python
np.random.rand(5) # random is a sub library from numpy, rand is the method, you specify the number and by default he creates
# random numbers between 0 and 1
```

```python
np.random.rand(5) * 100 # between 0 and 100
```

```python
np.random.rand(5, 3) * 100 # 2 dimension array (matrix) with random numbers between 0 and 100
```

rand provides a uniform distribution

```python
np.random.randn(4) # randn is no uniform like rand, its a Gausian distribution, its a normal distribution with mean 0 
# and standard deviation 1
```

```python
np.random.randint(0, 100, 10) # integers numbers
```

```python
np.round(np.random.rand(5)*100, 0) # equivalent to randint
```

```python
arr = np.random.rand(25) #creates a distribution
```

```python
arr #displays the vector of size 25 created above
```

```python
arr.reshape((5,5)) # reshapes the distribution to become a matrix and displays (he do not assign this values to arr)
```

```python
arr.shape # gonna display that still is a size 25 vector
```

```python
arr = arr.reshape((5,5)) # now assigning the reshape
```

```python
arr.shape # display that becomes a matrix
```

```python
arr.max() #max value
arr.min() #min value
arr.argmax() #max value position
arr.argmin() #min value position
```
