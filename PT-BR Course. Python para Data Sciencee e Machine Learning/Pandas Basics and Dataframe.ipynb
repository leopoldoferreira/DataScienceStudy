# Dataframe pt3
- Multilevel Indexes

```python
# Levels of Index
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1,2,3,1,2,3]
hier_index = list(zip(outside,inside)) # contains a list of tuples association each value from one list to another
hier_index = pd.MultiIndex.from_tuples(hier_index) # create multilevel index object
```

```python
df = pd.DataFrame(np.random.randn(6,2), index = hier_index, columns = ['A', 'B']) # tier dataframe
df
```

```python
df.loc['G1'].loc[1] # series of index 1
```

```python
df.index.names = ['Group', 'Number'] 
```

```python
df
```

```python
df.xs('G1') # xs = cross section
```

```python
df.xs(1, level = 'Number') # output indexes 1 of each group
```
