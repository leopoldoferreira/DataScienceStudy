**Create a basic fuction that return True if the world 'dog' contains on the input string.**

```python
def funcDog():
    findDog = 'Theres a dog there?'
    for item in findDog.split(' '):
        if item == 'dog':
            x = True
            return x
        else:
            x = False
    return x
funcDog()
```

**Finding in two lines**

```python
findDog = 'Theres a dog there?'.split(' ')
list(filter(lambda item:item=='dog', findDog))
```

**Create a function that contains the number of times the word 'dog' occurs in a string**

```python
def funcDog():
    countDog = 'This dog is faster then other dogs.'
    count = 0
    for item in countDog.split(' '):
        if 'dog' in item:
            count=count+1
    return count
funcDog()
```

**Encontrando as ocorrências em duas linhas**

```python
countDog = 'This dog is faster then other dogs.'.split(' ')
list(filter(lambda item:'dog' in item, countDog))
```

**Write a function that return 3 possibles outcomes: "No Ticket", "Small Ticket" ou "Big Ticket". <br>
If your speed is equal or inferior to 60 the output is No Ticket. If 61 and 80 (inclusive), the output is "MSmall Ticket". <br> 
81 or more, the result is "Big Ticket". Unless that is your birthday (coded with a boolean getted from a parameter in the function) <br> 
In your birthday, your velocity can be bigger in 5 on all cases**

```python
def velocity_capture(speed,is_birthday):
    if is_birthday == True:
        speed = speed - 5
    if speed <= 60:
        print('No Ticket')
    elif speed >=61 and speed <= 80:
        print('Small Ticket')
    elif speed >= 81:
        print('Big Ticket')
    print('Speed of {}km/h'.format(speed))
velocity_capture(65, False)
velocity_capture(65, True)
velocity_capture(81, False)
```
