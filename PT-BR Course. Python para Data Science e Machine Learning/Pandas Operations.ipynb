## Pandas Operations

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'col1':[1,2,3,4],'col2':[444,555,666,444],'col3':['abc','def','ghi','xyz']})
df.head() #show 
```

```python
df['col2'].unique() # only uniques values
df['col2'].nunique() # how many uniques values
df['col2'].value_counts() # unique and nunique together
```

```python
np.unique(df['col2']) # only uniques values in numpy
```

```python
df[(df['col1']>2) & (df['col2']==444)]
```

```python
def vezes2(x):
    return x*2
```

```python
df['col1'].sum() # sum all elements of col1
df['col1'].apply(vezes2) # like map. apply a method to each value
df['col3'].apply(len) # return length of each line from columns 3
```

```python
del df['col2'] # deletes
```

```python
df.sort_values(by='col2')
```

```python
df.isnul() # who is null?
```

```python
df.dropna() #drop NaN values 
```

```python
df['col1'].fillna(value=df['col1'].mean())
```
